```python
import random
import time
from IPython.display import Image, display

# --- INITIAL SETUP ---
while True:
    animal_type = input("What type of animal is it? (dog, cat, hamster): ").lower()
    if animal_type in ["dog", "cat", "hamster"]:
        break
    print("Please choose dog, cat, or hamster.")

name = input("What is your pet's name? ")

# Initial Stats
fullness, happiness, energy, money, cleanliness, health = 50, 50, 50, 50, 50, 100
food_level, toy_level, bed_level = 1, 1, 1

# Upgrade Names
food_names = {1: "pet food", 2: "vegetables", 3: "pizza", 4: "steak"}
toy_names = {1: "sock", 2: "ball", 3: "bone", 4: "play pen"}
bed_names = {1: "cardboard box", 2: "rug", 3: "bed", 4: "palace"}

def display_status():
    """Prints the status bar FIRST, then displays the pet image."""
    global fullness, happiness, energy, health, cleanliness, money
    
    # 1. CONSTRAIN STATS
    fullness = max(0, min(fullness, 100))
    happiness = max(0, min(happiness, 100))
    energy = max(0, min(energy, 100))
    health = max(0, min(health, 100))
    cleanliness = max(0, min(cleanliness, 100))

    # 2. PRINT STATUS BAR (Now First)
    print("\n================================")
    print(f"--- STATUS FOR {name.upper()} ---")
    print(f"Money: ${money}")
    print(f"Fullness: {fullness}/100 | Happiness: {happiness}/100")
    print(f"Energy: {energy}/100   | Health: {health}/100")
    print(f"Cleanliness: {cleanliness}/100")
    print(f"Equipment: {food_names[food_level]}, {toy_names[toy_level]}, {bed_names[bed_level]}")
    print("================================")
    print(f"Options: 1=Feed, 2=Play, 3=Sleep, 4=Shop, 5=Work, 6=Quit, 7=Clean")

    # 3. DISPLAY IMAGE (Now Second)
    img_path = ""
    try:
        if health < 30:
            img_path = "Sad dog.png" if animal_type == "dog" else f"sad {animal_type}.png"
        else:
            img_path = f"Neuteral {animal_type}.png" if animal_type != "hamster" else "neuteral hamster.png"
        display(Image(img_path, width=200))
    except:
        print(f"[Image {img_path} not found]\n")

# --- START GAME LOOP ---
while True:
    # This runs immediately after pressing 1, ensuring the first round is correct
    display_status()

    # CHECK FOR DEATH
    if fullness <= 0 or energy <= 0 or health <= 0:
        print(f"Oh no! {name} has passed away. Game Over.")
        break

    # PLAYER INPUT
    choice = input("What do you want to do? ")

    # PROCESS CHOICE
    if choice == "1":
        fullness += (food_level * 15)
        happiness -= 10; energy -= 10; cleanliness -= 10
        print(f"\n>> You fed {name} some {food_names[food_level]}!")

    elif choice == "2":
        happiness += (toy_level * 15)
        fullness -= 10; energy -= 10; cleanliness -= 10
        print(f"\n>> You played with the {toy_names[toy_level]}!")

    elif choice == "3":
        print("\n>> Sleeping...")
        time.sleep(1)
        energy += (bed_level * 20)
        fullness -= 10; happiness -= 10; cleanliness -= 5
        print(f">> {name} woke up feeling refreshed!")

    elif choice == "4":
        print("\n--- UPGRADE SHOP ---")
        if food_level < 4: print(f"1. Upgrade Food to {food_names[food_level+1]} ($50)")
        if toy_level < 4: print(f"2. Upgrade Toy to {toy_names[toy_level+1]} ($50)")
        if bed_level < 4: print(f"3. Upgrade Bed to {bed_names[bed_level+1]} ($50)")
        print("4. Exit Shop")
        
        shop_choice = input("Upgrade what? ")
        if shop_choice in ["1", "2", "3"] and money >= 50:
            if shop_choice == "1" and food_level < 4: 
                food_level += 1
                money -= 50
            elif shop_choice == "2" and toy_level < 4: 
                toy_level += 1
                money -= 50
            elif shop_choice == "3" and bed_level < 4: 
                bed_level += 1
                money -= 50
            print(">> Upgrade successful!")
        else:
            print(">> Action cancelled or insufficient funds.")

    elif choice == "5":
        work_choice = input("1. Easy Work ($15), 2. Hard Work ($45): ")
        if work_choice == "1":
            money += 15
            print(">> You earned $15.")
        else:
            money += 45
            print(">> Hard work pays off! You earned $45.")
        fullness -= 10; happiness -= 10; energy -= 15; cleanliness -= 10

    elif choice == "6":
        print("Goodbye!")
        break

    elif choice == "7":
        cleanliness = 100
        print(f"\n>> {name} is sparkling clean!")
        fullness -= 5; happiness -= 5; energy -= 5

    # RANDOM EVENTS
    if random.randint(1, 10) == 1:
        print("\n!!! VET VISIT REQUIRED !!!")
        vet_pay = input("Pay $100 for the vet? (1. Yes / 2. No): ")
        if vet_pay == "1" and money >= 100:
            money -= 100
            health = 100
            print(">> The vet healed your pet completely!")
        else:
            print(">> Health is dropping...")
            health -= 30

    print("\n--- NEW STATUS ---")
    time.sleep(1)
```
